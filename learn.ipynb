# **Temporal split function**

```python
def temporal_split(df, train_ratio=0.7, val_ratio=0.15):
    train_rows = []
    val_rows = []
    test_rows = []

    for user_id, user_df in df.groupby("user_idx"):
        user_df = user_df.sort_values("ts_ms")
        n = len(user_df)

        # Skip ultra-cold users
        if n < 3:
            continue

        train_end = int(n * train_ratio)
        val_end = int(n * (train_ratio + val_ratio))

        train_rows.append(user_df.iloc[:train_end])
        val_rows.append(user_df.iloc[train_end:val_end])
        test_rows.append(user_df.iloc[val_end:])

    train_df = pd.concat(train_rows)
    val_df = pd.concat(val_rows)
    test_df = pd.concat(test_rows)

    return train_df, val_df, test_df
train_df, val_df, test_df = temporal_split(final_df)
```

# **Metric implementations**

**Precision@K**

```python
def precision_at_k(recommended, ground_truth, k):
    if len(recommended) == 0:
        return 0
    return len(set(recommended[:k]) & set(ground_truth)) / k
```

**Recall@K**

```python
def recall_at_k(recommended, ground_truth, k):
    if len(ground_truth) == 0:
        return 0
    return len(set(recommended[:k]) & set(ground_truth)) / len(ground_truth)
```

**NDCG@K**

```python
import numpy as np

def ndcg_at_k(recommended, ground_truth, k):
    dcg = 0.0
    for i, item in enumerate(recommended[:k]):
        if item in ground_truth:
            dcg += 1 / np.log2(i + 2)
    idcg = sum(1 / np.log2(i + 2) for i in range(min(len(ground_truth), k)))
    return dcg / idcg if idcg > 0 else 0
```

**MRR**

```python
def mrr(recommended, ground_truth):
    for i, item in enumerate(recommended):
        if item in ground_truth:
            return 1 / (i + 1)
    return 0
```

# **Collaborative filtering**

```python
# Re-encode users
train_df["user_idx_cf"] = train_df["user_idx"].astype("category").cat.codes

# Re-encode items
train_df["item_idx_cf"] = train_df["item_idx"].astype("category").cat.codes
```

```python
from scipy.sparse import coo_matrix

n_users = train_df["user_idx_cf"].nunique()
n_items = train_df["item_idx_cf"].nunique()

user_item_matrix = coo_matrix(
    (
        train_df["interaction_strength"],
        (train_df["user_idx_cf"], train_df["item_idx_cf"])
    ),
    shape=(n_users, n_items)
)
```

```python
user_mapping = (
    train_df[["user_idx", "user_idx_cf"]]
    .drop_duplicates()
    .set_index("user_idx")["user_idx_cf"]
    .to_dict()
)

item_mapping = (
    train_df[["item_idx", "item_idx_cf"]]
    .drop_duplicates()
    .set_index("item_idx")["item_idx_cf"]
    .to_dict()
)
```

```python
test_df = test_df[test_df["user_idx"].isin(user_mapping)]
test_df = test_df[test_df["item_idx"].isin(item_mapping)]

test_df["user_idx_cf"] = test_df["user_idx"].map(user_mapping)
test_df["item_idx_cf"] = test_df["item_idx"].map(item_mapping)
```

```python
from implicit.als import AlternatingLeastSquares

model = AlternatingLeastSquares(
    factors=64,
    regularization=0.1,
    iterations=20,
    random_state=42
)

# implicit expects ITEM x USER matrix
model.fit(user_item_matrix.T)
```

```python
from scipy.sparse import csr_matrix

user_item_csr = user_item_matrix.tocsr()

def recommend_cf(user_idx_cf, k=10):
    user_items = user_item_csr[user_idx_cf]
    recs, scores = model.recommend(
        user_idx_cf,
        user_items,
        N=k,
        filter_already_liked_items=True
    )
    return recs.tolist()
```

```python
valid_cf_users = set(train_df["user_idx_cf"].unique())
```

```python
test_user_items_cf = (
    test_df[test_df["user_idx_cf"].isin(valid_cf_users)]
    .groupby("user_idx_cf")["item_idx_cf"]
    .apply(list)
    .to_dict()
)
```

```python
def evaluate_cf(k=10):
    precisions, recalls, ndcgs, mrrs = [], [], [], []

    for user_idx_cf, true_items in test_user_items_cf.items():
        if user_idx_cf >= model.user_factors.shape[0]:
            continue  # safety guard

        recs = recommend_cf(user_idx_cf, k)

        precisions.append(precision_at_k(recs, true_items, k))
        recalls.append(recall_at_k(recs, true_items, k))
        ndcgs.append(ndcg_at_k(recs, true_items, k))
        mrrs.append(mrr(recs, true_items))

    return {
        "Precision@K": np.mean(precisions),
        "Recall@K": np.mean(recalls),
        "NDCG@K": np.mean(ndcgs),
        "MRR": np.mean(mrrs)
    }
```

# **Sequence Model (GRU4Rec-style)**

```python
train_sequences = (
    train_df
    .sort_values(["user_idx", "ts_ms"])
    .groupby("user_idx")["item_idx"]
    .apply(list)
)
```

```python
X, y = [], []

for seq in train_sequences:
    if len(seq) < 2:
        continue
    for i in range(1, len(seq)):
        X.append(seq[:i])
        y.append(seq[i])
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

max_len = 20  # cap history length
X_padded = pad_sequences(X, maxlen=max_len, padding="pre")
y = np.array(y)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, GRU, Dense

n_items = train_df["item_idx"].nunique()

inputs = Input(shape=(max_len,))
x = Embedding(input_dim=n_items + 1, output_dim=64, mask_zero=True)(inputs)
x = GRU(64)(x)
outputs = Dense(n_items, activation="softmax")(x)

model = Model(inputs, outputs)
model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
model.fit(
    X_padded,
    y,
    epochs=20,
    batch_size=256,
    validation_split=0.1
)
```

```python
model.save("models/gru_model.h5")
```

```python
model.save("models/gru_model.keras")
```

```python
def recommend_sequence(user_history, k=10):
    padded = pad_sequences([user_history], maxlen=max_len, padding="pre")
    probs = model.predict(padded, verbose=0)[0]
    return np.argsort(probs)[-k:][::-1]
```

**Evaluate on TEST set**

```python
test_user_items = (
    test_df.groupby("user_idx")["item_idx"]
    .apply(list)
    .to_dict()
)

def evaluate_sequence(k=10):
    precisions, recalls, ndcgs, mrrs = [], [], [], []

    for user_idx, true_items in test_user_items.items():
        history = (
            train_df[train_df.user_idx == user_idx]["item_idx"].tolist() +
            val_df[val_df.user_idx == user_idx]["item_idx"].tolist()
        )

        if len(history) == 0:
            continue

        recs = recommend_sequence(history, k)

        precisions.append(precision_at_k(recs, true_items, k))
        recalls.append(recall_at_k(recs, true_items, k))
        ndcgs.append(ndcg_at_k(recs, true_items, k))
        mrrs.append(mrr(recs, true_items))

    return {
        "Precision@K": np.mean(precisions),
        "Recall@K": np.mean(recalls),
        "NDCG@K": np.mean(ndcgs),
        "MRR": np.mean(mrrs)
    }
```

```python
seq_metrics = evaluate_sequence(k=10)
seq_metrics
```

# **Hybrid model**

```python
from collections import defaultdict

user_histories = defaultdict(list)

# Use TRAIN data
for row in train_df.sort_values("ts_ms").itertuples():
    user_histories[row.user_idx].append(row.item_idx)

# Add VALIDATION data
for row in val_df.sort_values("ts_ms").itertuples():
    user_histories[row.user_idx].append(row.item_idx)
```

```python
def hybrid_recommend(user_idx, k=10):
    history = user_histories.get(user_idx, [])

    if len(history) == 0:
        return recommend_most_popular(user_idx, k)

    if len(history) < 3:
        pop = recommend_most_popular(user_idx, k)
        seq = recommend_sequence(history, k)
        return list(dict.fromkeys(seq + pop))[:k]

    return recommend_sequence(history, k)
```

```python
def evaluate_hybrid(k=10):
    precisions, recalls, ndcgs, mrrs = [], [], [], []

    for user_idx, true_items in test_user_items.items():
        recs = hybrid_recommend(user_idx, k)

        precisions.append(precision_at_k(recs, true_items, k))
        recalls.append(recall_at_k(recs, true_items, k))
        ndcgs.append(ndcg_at_k(recs, true_items, k))
        mrrs.append(mrr(recs, true_items))

    return {
        "Precision@K": np.mean(precisions),
        "Recall@K": np.mean(recalls),
        "NDCG@K": np.mean(ndcgs),
        "MRR": np.mean(mrrs)
    }
evaluate_hybrid()
```

# **candidate**

```python
def generate_candidates(user_idx, k=100):
    candidates = set()

    candidates.update(popular_items[:50])

    if user_idx in user_histories:
        history = user_histories[user_idx]
        candidates.update(history[-10:])

    return list(candidates)[:k]
```

```python
def evaluate_candidate_generation(max_candidates=100):
    hits = 0
    total = 0

    for user_idx, true_items in test_user_items.items():
        candidates = generate_candidates(user_idx, max_candidates)

        # Check if ANY true item is in candidates
        if any(item in candidates for item in true_items):
            hits += 1
        total += 1

    return hits / total
```

```python
candidate_recall = evaluate_candidate_generation(max_candidates=100)
candidate_recall
```

# **rank pipeline**

```python
def rank_candidates(user_history, candidates, k=10):
    padded = pad_sequences([user_history], maxlen=max_len)
    scores = model.predict(padded)[0]

    ranked = sorted(
        candidates,
        key=lambda x: scores[x],
        reverse=True
    )
    return ranked[:k]
```

```python
def hybrid_rank(user_idx, k=10):
    # Step 1: Candidate generation
    candidates = generate_candidates(user_idx)

    # Step 2: Ranking
    return rank_candidates(user_idx, candidates, k)
```

```python
def evaluate_ranking(k=10):
    precisions, recalls, ndcgs, mrrs = [], [], [], []

    for user_idx, true_items in test_user_items.items():
        recs = hybrid_rank(user_idx, k)

        precisions.append(precision_at_k(recs, true_items, k))
        recalls.append(recall_at_k(recs, true_items, k))
        ndcgs.append(ndcg_at_k(recs, true_items, k))
        mrrs.append(mrr(recs, true_items))

    return {
        "Precision@K": np.mean(precisions),
        "Recall@K": np.mean(recalls),
        "NDCG@K": np.mean(ndcgs),
        "MRR": np.mean(mrrs)
    }
```

```python
ranking_metrics = evaluate_ranking(k=10)
ranking_metrics
```

# **Coverage**

```python
def compute_coverage(recommend_fn, users, k=10):
    recommended_items = set()

    for u in users:
        recs = recommend_fn(u, k)
        recommended_items.update(recs)

    total_items = train_df["item_idx"].nunique()
    return len(recommended_items) / total_items
```

```python
users = list(test_user_items.keys())

coverage_pop = compute_coverage(recommend_most_popular, users)
coverage_hybrid = compute_coverage(hybrid_recommend, users)

coverage_pop, coverage_hybrid
```
